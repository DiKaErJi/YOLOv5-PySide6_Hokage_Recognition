```python
import cv2
import matplotlib.pyplot as plt
```

```python
# 打开视频文件
video = cv2.VideoCapture("./BVN.mp4")

# 读取一帧
# 如果顺利读到了下一帧则ret返回True，frame是一帧的内容
ret, frame = video.read()
```

```python
plt.imshow(frame)
```

```python
# 通道转换，从BGR转到RGB
plt.imshow(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
```

```python
video =cv2.VideoCapture("./BVN.mp4")
num=0
save_step=30
while True:
    ret,frame = video.read()
    if not ret:
        break
    num+=1
    if num%save_step==0:
        cv2.imwrite("./images/"+str(num)+".jpg",frame)
```
